```python
import pandas as pd
import collections

DATA_DIR = '../data/'

with open(DATA_DIR + 'results.json', 'r') as file:
    results = pd.read_json(file, orient='index')
```

```python
results.head()
```

```python
def make_counter(value):
    text = results.loc[results['value change']==value]
    words = []
    for text in text['text']:
        sanat = text.split(' ')
        for sana in sanat:
            words.append(sana)
    return collections.Counter(words).most_common(100)
```

```python
bigpos = make_counter('BIGPOS')
bigneg = make_counter('BIGNEG')
```

```python
for i in range(0,100):
    print(bigpos[i], ' : ', bigneg[i])
```
